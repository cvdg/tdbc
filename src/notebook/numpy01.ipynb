```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def f(x):
   return x ** 2 - 3 * x + 6
```

```python
x = np.linspace(-5, 5, 1000)

df = pd.DataFrame()
df['x'] = np.linspace(-5, 5, 1000)
df['y'] = f(df['x'])

df.tail()
```

```python
sns.lineplot(df, x='x', y='y')
plt.show()
```
