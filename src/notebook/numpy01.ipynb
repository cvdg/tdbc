A simple test with a basic mathematical formula and plot.

Setup the needed modules.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')
```

A simple function:

$$y = ax^2 + bx + c$$

```python
a = 1
b = -2
c = -6
```

```python
def f(x):
   return a * x ** 2 + b * x + c
```

Create the data.

* __x__: array from -5 to 5 in steps of 0.01
* __y__: array of f(x)

* __df__: the dataframe with the results

Plot the result.

```python
x = np.linspace(-5, 5, 1000)

df = pd.DataFrame()
df['x'] = np.linspace(-5, 5, 1000)
df['y'] = f(df['x'])

sns.lineplot(df, x='x', y='y')
plt.show()
```
